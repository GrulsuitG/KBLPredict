```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

RANDOM_SEED = 2022
```

```python
from sklearn.model_selection import train_test_split

data = pd.read_csv('/content/drive/MyDrive/KBL Prediction/dataset/ver3_10game_2nd_Data.csv')
data
```

```python
data.drop(columns=["win"], inplace=True)
df = data.iloc[:, 1:]
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=RANDOM_SEED)
```

```python
X_train, y_train
```

```python
from autosklearn import regression, metrics
model = regression.AutoSklearnRegressor(
    time_left_for_this_task=60*10,
    n_jobs=-1,
    metric=metrics.r2,
    resampling_strategy='cv',
    resampling_strategy_arguments={'folds': 3},
    seed=RANDOM_SEED
)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.leaderboard(detailed=True, ensemble_only=False)
```

```python
#performance over time
import matplotlib.pyplot as plt
model.performance_over_time_.plot(
        x='Timestamp',
        kind='line',
        legend=True,
        title='Auto-sklearn accuracy over time',
        grid=True,)
plt.show()
```
